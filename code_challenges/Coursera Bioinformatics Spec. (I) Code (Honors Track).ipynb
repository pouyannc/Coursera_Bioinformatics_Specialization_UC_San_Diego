```python
def PatternCount(text,pattern):
    '''outputs the number of times pattern appears in text'''
    count = 0
    for i in range(len(text)-len(pattern)):
        if text[i:i+len(pattern)] == pattern:
            count += 1
    return count
    
with open('Downloads/dataset_2_7 (3).txt') as handle:
    data = [x for x in handle.read().split()]

print data
print PatternCount(data[0],data[1])
```

```python
def FrequentPatterns(text, k):
    '''outputs the most frequent k-mer in text'''
    freq_patterns = []
    count=[]
    n = len(text)-k
    for i in range(n):
        pattern = text[i:i+k]
        count.append(PatternCount(text, pattern))
    max_count = max(count)
    for i in range(n):
        if count[i] == max_count:
            freq_patterns.append(text[i:i+k])
    return " ".join(list(dict.fromkeys(freq_patterns)))
        
with open('Downloads/dataset_2_10.txt') as handle:
    data = [x for x in handle.read().split()]

print data
print FrequentPatterns(data[0],int(data[1]))
```

```python
def Complement(string):
    '''returns the reverse complement of string'''
    reverse = string[::-1]
    reverse = reverse.replace("A","Ax")
    reverse = reverse.replace("C","Cx")
    reverse = reverse.replace("G","C")
    reverse = reverse.replace("T","A")
    reverse = reverse.replace("Ax","T")
    reverse = reverse.replace("Cx","G")
    return reverse

with open('Downloads/dataset_3_2.txt') as handle:
    data = [x for x in handle.read().split()]
    
print data
print Complement(data[0])
```

```python
from __future__ import print_function

def FindMotif(substring,string):
    '''returns the locations where substring appears in string'''
    n = len(string)
    m = len(substring)
    motif_positions = []
    for i in range(n-m):
        if string[i:i+m] == substring:
            motif_positions.append(i)
    return motif_positions

with open('Downloads/Vibrio_cholerae.txt') as handle:
    data = [x for x in handle.read().split()]
    
print (data)
answer = FindMotif('CTTGATCAT',data[0])
print (*answer, sep = ' ')
```

```python
def ClumpFinder(genome,k,L,t):
    '''returns the distinct k-mers which appear t times in each window of size L in the genome'''
    patterns = []
    n = len(genome)-L+1
    m = L-k+1
    for i in range(n):
        for j in range(m):
            
            if (PatternCount(genome[i:i+L], genome[j+i:j+i+k])) == t:
                patterns.append(genome[j+i:j+i+k])
    return " ".join(list(dict.fromkeys(patterns)))


with open('Downloads/dataset_4_5.txt') as handle:
    data = [x for x in handle.read().split()]
    
print (data)
print (ClumpFinder(data[0],int(data[1]),int(data[2]),int(data[3])))
```

```python
def ComputingFrequencies(text, k):
    '''returns a list of k-mer frequencies -- indices indicate the particular k-mer (based on PatternToNumber)'''
    n = len(text)-k
    frequencies = []
    pattern =''
    for i in range(4**k):
        frequencies.append(0)
    for i in range(n+1):
        pattern = text[i:i+k]
        j = PatternToNumber(pattern)
        frequencies[j] += 1
    return frequencies


def PatternToNumber(pattern):
    '''returns the index of a k-mer between 0 and (4^k)-1'''
    base_values = {'A':0,'C':1,'G':2,'T':3}
    k = len(pattern)
    counter =0 
    number = 0
    for base in pattern:
        counter+=1
        number += base_values[base] * (4**(k-counter))
    return number

with open('Downloads/dataset_2994_5.txt') as handle:
    data = [x for x in handle.read().split()]

answer = ComputingFrequencies(data[0], int(data[1]))
#print (data)
print (*answer, sep= ' ')
```
